```python
import pandas as pd
from glob import glob
```

```python
pd.set_option('display.max_rows', 1000)
```

```python
dfs = []
for f in glob('matches/*.csv'):
    team = pd.read_csv(f)
    filename = f.split('/')[-1]
    segment = filename[-5]
    team['team'] = filename[0:3] if segment == '1' else filename[4:7]
    team['match'] = filename[0:7]
    dfs.append(team)
df = pd.concat(dfs, axis=0).reset_index(drop=True)
df.head()
```

```python
df.columns.tolist()
```

```python
df['xG'].sum(), df['Gls'].sum()
```

```python
sums = df.groupby(['Player', '#', 'team']).sum().reset_index()
sums = sums[sums['Min'] > 90]
sums
```

```python
sums['npxG_per_90'] = sums.apply(lambda row: row['npxG'] / row['Min'] * 90, axis=1)
sums.sort_values(by='npxG_per_90')[['Player', 'team', 'npxG', 'Min', 'Gls', 'PK', 'npxG_per_90']].tail(20)
```

```python
sums['gls_per_npxG'] = sums.apply(lambda row: 0 if row['npxG'] == 0 or row['Gls']-row['PK'] < 1 else (row['Gls']-row['PK']) / row['npxG'], axis=1)
gls_per_npxG = sums.sort_values(by='gls_per_npxG')[['Player', 'team', 'npxG', 'Min', 'Gls', 'PK', 'gls_per_npxG']]
gls_per_npxG = gls_per_npxG[gls_per_npxG['Gls'] - gls_per_npxG['PK'] >= 1]
print(gls_per_npxG.shape[0])
gls_per_npxG
```
