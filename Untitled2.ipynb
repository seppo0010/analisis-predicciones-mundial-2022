```python
import pandas as pd
import numpy as np
import re
import json
```

```python
results = pd.read_csv('results.csv')[['Round', 'Home', 'Away', 'Score']]
fivethirtyeight = pd.read_csv('538.csv').rename(columns={
    'proba': '538_proba',
    'probx': '538_probx',
    'probb': '538_probb',
}).drop(['teama', 'teamb'], axis=1)
m301060 = pd.read_csv('301060.csv')
m301060.columns = ['proba', 'teama', 'probx', 'teamb', 'probb']
m301060['proba'] /= 100
m301060['probx'] /= 100
m301060['probb'] /= 100
m301060 = m301060.rename(columns={
    'proba': '301060_proba',
    'probx': '301060_probx',
    'probb': '301060_probb',
}).drop(['teama', 'teamb'], axis=1)
names = {
    'Qatar': 'qat',
    'Ecuador': 'ecu',
    'England': 'eng',
    'Senegal': 'sen',
    'United States': 'usa',
    'Argentina': 'arg',
    'Denmark': 'den',
    'Mexico': 'mex',
    'France': 'fra',
    'Morocco': 'mor',
    'Germany': 'ger',
    'Spain': 'spa',
    'Belgium': 'bel',
    'Switzerland': 'swi',
    'Uruguay': 'uru',
    'Portugal': 'por',
    'Brazil': 'bra',
    'Wales': 'wal',
    'Netherlands': 'ned',
    'Tunisia': 'tun',
    'Poland': 'pol',
    'Japan': 'jap',
    'Croatia': 'cro',
    'Cameroon': 'cam',
    'Korea Republic': 'kor',
    'Ecuador': 'ecu',
    'Iran': 'ira',
    'Australia': 'aus',
    'Saudi Arabia': 'sau',
    'Canada': 'can',
    'Costa Rica': 'cos',
    'Ghana': 'gha',
    'Serbia': 'ser',
}

def read_oddspedia(f):
    with open(f'oddspedia/{f}.json', 'r') as fp:
        data = json.loads(fp.read())['data']['prematch'][0]['periods'][0]['odds']
    series = {}
    for row in data:
        p1, p2, p3 = 1/float(row['o1']), 1/float(row['o2']), 1/float(row['o3'])
        series[f'{row["bookie_slug"]}_proba'] = p1/(p1+p2+p3)
        series[f'{row["bookie_slug"]}_probx'] = p2/(p1+p2+p3)
        series[f'{row["bookie_slug"]}_probb'] = p3/(p1+p2+p3)
    return pd.Series(series)

oddspedia = results.apply(lambda row: read_oddspedia(f'{names[row["Home"]]}-{names[row["Away"]]}'), axis=1)
df = results.join(fivethirtyeight).join(m301060).join(oddspedia.dropna(axis=1))
df
```

```python
def remove_tie(row):
    if row['Round'] != 'Group stage':
        for key, value in row.items():
            if key.endswith('_probx') and value > 0:
                row[key] = 0.0
                row[key[:-1] + 'a'] += value / 2
                row[key[:-1] + 'b'] += value / 2
    return pd.Series(row)
df = df.apply(remove_tie, axis=1)
df
```

```python
def winner(score):
    if isinstance(score, float) and np.isnan(score):
        return None
    scores = score.split(' ')
    if len(scores) == 1:
        a, b = int(score.split('–')[0]), int(score.split('–')[1])
        if a == b:
            return 'x'
        if a > b:
            return 'a'
        return 'b'
    segs = re.split(r'(\d+)', score)
    if int(segs[1]) < int(segs[7]):
        return 'b'
    return 'a'
df['winner'] = df['Score'].apply(winner)
df
```

```python
providers = [x[:-6] for x in df.columns.tolist() if x.endswith('_proba')]
providers
```

```python
def brier_score_row(row, prefix):
    return (
        pow(row[f'{prefix}_proba'] - (1 if row['winner'] == 'a' else 0), 2) +
        pow(row[f'{prefix}_probb'] - (1 if row['winner'] == 'b' else 0), 2) +
        pow(row[f'{prefix}_probx'] - (1 if row['winner'] == 'x' else 0), 2)
    )
for p in providers:
    df[f'brier_score_{p}'] = df.apply(lambda row: brier_score_row(row, p), axis=1)
df
```

```python
df['dif'] = df.apply(lambda row: (row['538_proba'] > row['538_probb']) != (row['301060_proba'] > row['301060_probb']), axis=1)
df[df['dif']][['Home', 'Away', 'Score',
                          '538_proba', '538_probx', '538_probb', 'brier_score_538',
                          '301060_proba', '301060_probx', '301060_probb', 'brier_score_301060',
                         ]]
```

```python
df['surprise'] = df.apply(lambda row: sum([v for k, v in row.items() if 'brier_score_' in k]), axis=1)
df.sort_values(by='surprise')
```
