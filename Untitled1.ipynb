```python
import pandas as pd
from glob import glob
```

```python
df = pd.concat([pd.read_csv(f) for f in glob('matches/*.csv')], axis=0).reset_index(drop=True)
df
```

```python
df.columns.tolist()
```

```python
df.groupby('Player').sum().apply(lambda row: (row['SoT']) / row['Min'] * 90 if row['SoT'] >= 2 else 0, axis=1).sort_values().tail(20)
```

```python
df.groupby('Player').sum().apply(lambda row: (row['SoT']), axis=1).sort_values().tail(20)
```
