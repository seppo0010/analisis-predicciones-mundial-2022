```python
import pandas as pd
import numpy as np
import re
import json
```

```python
fivethirtyeight = pd.read_csv('538.csv')
fivethirtyeight.head()
```

```python
results = pd.read_csv('results.csv')
results.head()
```

```python
df = fivethirtyeight.join(results)
def winnera(score, penalties=True):
    if isinstance(score, float) and np.isnan(score):
        return False
    scores = score.split(' ')
    if len(scores) == 1:
        return int(score.split('–')[0]) > int(score.split('–')[1])
    if penalties == False:
        return False
    segs = re.split(r'(\d+)', score)
    return int(segs[1]) > int(segs[7])

def winnerb(score, penalties=True):
    if isinstance(score, float) and np.isnan(score):
        return False
    scores = score.split(' ')
    if len(scores) == 1:
        return int(score.split('–')[0]) < int(score.split('–')[1])
    if penalties == False:
        return False
    segs = re.split(r'(\d+)', score)
    return int(segs[1]) < int(segs[7])

def tie(score, penalties=True):
    if isinstance(score, float) and np.isnan(score):
        return False
    scores = score.split(' ')
    if len(scores) == 1:
        return int(score.split('–')[0]) == int(score.split('–')[1])
    if penalties == False:
        return True
    return False

df['proba_win'] = df['proba'] * df['Score'].apply(winnera)
df['probb_win'] = df['probb'] * df['Score'].apply(winnerb)
df['probx_win'] = df['probx'] * df['Score'].apply(tie)
print(df['proba_win'].sum() + df['probb_win'].sum() + df['probx_win'].sum())
df[['teama', 'teamb', 'proba_win', 'probx_win', 'probb_win', 'Score']].iloc[57]
```

```python
def goals(score):
    if isinstance(score, float) and np.isnan(score):
        return 0
    scores = score.split(' ')
    if len(scores) == 1:
        return int(score.split('–')[0]) + int(score.split('–')[1])
    segs = re.split(r'(\d+)', score)
    return int(segs[3]) + int(segs[5])

df['Score'].apply(goals).sum()
```

```python
def is_close(score):
    if isinstance(score, float) and np.isnan(score):
        return 0
    scores = score.split(' ')
    if len(scores) == 1:
        return abs(int(score.split('–')[0]) - int(score.split('–')[1])) < 2
    return True

df['Score'].apply(is_close).sum()
```

```python
names = {
    'Qatar': 'qat',
    'Ecuador': 'ecu',
    'England': 'eng',
    'Senegal': 'sen',
    'United States': 'usa',
    'Argentina': 'arg',
    'Denmark': 'den',
    'Mexico': 'mex',
    'France': 'fra',
    'Morocco': 'mor',
    'Germany': 'ger',
    'Spain': 'spa',
    'Belgium': 'bel',
    'Switzerland': 'swi',
    'Uruguay': 'uru',
    'Portugal': 'por',
    'Brazil': 'bra',
    'Wales': 'wal',
    'Netherlands': 'ned',
    'Tunisia': 'tun',
    'Poland': 'pol',
    'Japan': 'jap',
    'Croatia': 'cro',
    'Cameroon': 'cam',
    'Korea Republic': 'kor',
    'Ecuador': 'ecu',
    'Iran': 'ira',
    'Australia': 'aus',
    'Saudi Arabia': 'sau',
    'Canada': 'can',
    'Costa Rica': 'cos',
    'Ghana': 'gha',
    'Serbia': 'ser',
}

def read_oddspedia(f):
    if f == 'arg-fra': return None
    with open(f'oddspedia/{f}.json', 'r') as fp:
        data = json.loads(fp.read())['data']['prematch'][0]['periods'][0]['odds']
    series = {}
    for row in data:
        p1, p2, p3 = 1/float(row['o1']), 1/float(row['o2']), 1/float(row['o3'])
        series[f'{row["bookie_slug"]}_proba'] = p1/(p1+p2+p3)
        series[f'{row["bookie_slug"]}_probx'] = p2/(p1+p2+p3)
        series[f'{row["bookie_slug"]}_probb'] = p3/(p1+p2+p3)
    return pd.Series(series)

oddspedia = results.apply(lambda row: read_oddspedia(f'{names[row["Home"]]}-{names[row["Away"]]}'), axis=1)
oddspedia
```

```python
df = oddspedia.join(results)
df['proba_win'] = df['bet365_proba'] * df['Score'].apply(lambda score: winnera(score, penalties=False))
df['probb_win'] = df['bet365_probb'] * df['Score'].apply(lambda score: winnerb(score, penalties=False))
df['probx_win'] = df['bet365_probx'] * df['Score'].apply(lambda score: tie(score, penalties=False))
print(df['proba_win'].sum() + df['probb_win'].sum() + df['probx_win'].sum())
df[['Home', 'Away', 'proba_win', 'probx_win', 'probb_win', 'Score']].iloc[57]
```
