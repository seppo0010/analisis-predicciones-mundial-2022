```python
import pandas as pd
import numpy as np
import re
import json
```

```python
results = pd.read_csv('results-2018.csv')
results['Home'] = results['Home'].apply(lambda s: ' '.join(s.split(' ')[:-1]))
results['Away'] = results['Away'].apply(lambda s: ' '.join(s.split(' ')[1:]))
results = results[['Round', 'Home', 'Away', 'Score']].set_index(['Home', 'Away'])
fivethirtyeight = pd.read_csv('538-2018.csv').rename(columns={
    'proba': '538_proba',
    'probx': '538_probx',
    'probb': '538_probb',
})
fivethirtyeight['teama'] = fivethirtyeight['teama'].apply(lambda x: x if x != 'USA' else 'United States')
fivethirtyeight['teamb'] = fivethirtyeight['teamb'].apply(lambda x: x if x != 'USA' else 'United States')
fivethirtyeight['teama'] = fivethirtyeight['teama'].apply(lambda x: x if x != 'South Korea' else 'Korea Republic')
fivethirtyeight['teamb'] = fivethirtyeight['teamb'].apply(lambda x: x if x != 'South Korea' else 'Korea Republic')
fivethirtyeight['teama'] = fivethirtyeight['teama'].apply(lambda x: x if x != 'Iran' else 'IR Iran')
fivethirtyeight['teamb'] = fivethirtyeight['teamb'].apply(lambda x: x if x != 'Iran' else 'IR Iran')
fivethirtyeight = fivethirtyeight.rename(columns={'teama': 'Home', 'teamb': 'Away'}).set_index(['Home', 'Away'])
df = results.join(fivethirtyeight)
df
```

```python
def winner(score):
    if isinstance(score, float) and np.isnan(score):
        return None
    scores = score.split(' ')
    if len(scores) == 1:
        a, b = int(score.split('–')[0]), int(score.split('–')[1])
        if a == b:
            return 'x'
        if a > b:
            return 'a'
        return 'b'
    segs = re.split(r'(\d+)', score)
    if int(segs[1]) < int(segs[7]):
        return 'b'
    return 'a'
df['winner'] = df['Score'].apply(winner)
df
```

```python
providers = ['538']
def brier_score_row(row, prefix):
    return (
        pow(row[f'{prefix}_proba'] - (1 if row['winner'] == 'a' else 0), 2) +
        pow(row[f'{prefix}_probb'] - (1 if row['winner'] == 'b' else 0), 2) +
        pow(row[f'{prefix}_probx'] - (1 if row['winner'] == 'x' else 0), 2)
    )
for p in providers:
    df[f'brier_score_{p}'] = df.apply(lambda row: brier_score_row(row, p), axis=1)
df
```

```python
{p: (1-df[f'brier_score_{p}'].sum()/40) for p in providers}
```

```python
df['surprise'] = df.apply(lambda row: sum([v for k, v in row.items() if 'brier_score_' in k]), axis=1)
df.sort_values(by='surprise')
```
